```python
import pandas as pd
datos = pd.read_csv("varones_20_anios_peso_saludable_10000.csv")
```

```python
import seaborn as sb


sb.scatterplot(x="Altura", y="Peso", data=datos,
               hue="Peso", palette="coolwarm")
```

```python
#Caracteristicas (X), etiqueta (y)
X = datos["Altura"]
y = datos["Peso"]
```

```python
X_p = X.values.reshape(-1,1)
y_p = y.values.reshape(-1,1)
```

```python
# Creamos el modelo
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
#Entrenamiento
modelo.fit(X_p, y_p)
```

```python
# Predicciones del modelo
altura = 187
prediccion = modelo.predict([[altura]])
resultado = round(prediccion.item(), 2)
print(f"Con {altura} cm pesaras aprox {resultado} kg")
```

```python
# El grado de error, nunca ser cero cuando hay datos menos exactos
modelo.score(X_p, y_p)
```
